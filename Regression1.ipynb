```python
import numpy as np
import pandas as pd
from sklearn.model_selection import KFold
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error

# Generate approximate latitude & longitude (since Boston_df doesn't have them)
np.random.seed(42)
Boston_df["latitude"] = np.random.uniform(42.23, 42.42, size=len(Boston_df))
Boston_df["longitude"] = np.random.uniform(-71.20, -70.95, size=len(Boston_df))

# Define KFold cross-validation
num_splits = 5
num_models = 4
kfold = KFold(n_splits=num_splits, random_state=216, shuffle=True)

# Create an array to store RMSE values
rmses = np.zeros((num_models, num_splits))

# Define RMSE function
def root_mean_squared_error(y_true, y_pred):
    return np.sqrt(mean_squared_error(y_true, y_pred))

# Loop through KFold splits
for i, (train_index, test_index) in enumerate(kfold.split(Boston_df)):

    # Split dataset into training and testing
    train_set = Boston_df.iloc[train_index]
    test_set = Boston_df.iloc[test_index]

    # Target variable
    y_test = test_set["medv"]

    ## 1️⃣ Baseline Model (Mean Prediction)
    baseline_pred = np.full(len(test_set), train_set["medv"].mean())
    rmses[0, i] = root_mean_squared_error(y_test, baseline_pred)

    ## 2️⃣ Simple Linear Regression (SLR) - Using 'rm' (average number of rooms)
    slr = LinearRegression()
    slr.fit(train_set[['rm']], train_set["medv"])
    rmses[1, i] = root_mean_squared_error(y_test, slr.predict(test_set[['rm']]))

    ## 3️⃣ Multiple Linear Regression (MLR) - Using 'rm' & 'lstat'
    mlr = LinearRegression()
    mlr.fit(train_set[['rm', 'lstat']], train_set["medv"])
    rmses[2, i] = root_mean_squared_error(y_test, mlr.predict(test_set[['rm', 'lstat']]))

    ## 4️⃣ K-Nearest Neighbors (KNN) Regression - Using 'latitude' & 'longitude'
    knn = KNeighborsRegressor(n_neighbors=10)
    knn.fit(train_set[['latitude', 'longitude']], train_set["medv"])
    rmses[3, i] = root_mean_squared_error(y_test, knn.predict(test_set[['latitude', 'longitude']]))

# Display the RMSE results
print("RMSE values for each model across folds:\n", rmses)
print("\nMean RMSE for each model:\n", np.mean(rmses, axis=1))
```

```python
# Model names for visualization
model_names = ["Baseline", "SLR (rm)", "MLR (rm, lstat)", "KNN (lat, lon)"]

# Compute mean RMSE for each model
mean_rmse = np.mean(rmses, axis=1)

# 📊 OPTION 1: Bar Chart - Mean RMSE per Model
plt.figure(figsize=(8, 5))
sns.barplot(x=model_names, y=mean_rmse, hue=model_names, dodge=False, palette="viridis", legend=False)
plt.ylabel("Mean RMSE")
plt.title("Model Performance Comparison (Lower is Better)")
plt.xticks(rotation=15)
plt.show()

# 📊 OPTION 2: Box Plot - RMSE Distribution Across Folds
plt.figure(figsize=(8, 5))
sns.boxplot(data=rmses.T, palette="coolwarm")
plt.xticks(ticks=range(len(model_names)), labels=model_names, rotation=15)
plt.ylabel("RMSE")
plt.title("RMSE Distribution Across K-Folds")
plt.show()




```
